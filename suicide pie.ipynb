```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('master.csv')
```

```python
data.head()
```

```python
bar_data = data.groupby('age')['suicides_no'].sum().to_dict()
bar_data
```

```python
labels = '15-24 years', '25-34 years', '35-54 years', '5-14 years', '55-74 years', '75+ years'
sizes = [808542, 1123912, 2452141, 52264, 1658443, 653118]
explode = [0, 0, 0, 100, 0, 0]
plt.pie(sizes, explode=explode, labels=labels, 
autopct='%1.1f%%', shadow=False, radius= 100, startangle=-90, pctdistance=0.7)

plt.axis('equal')
plt.title('Total # of Suicides of Each Age Range')
plt.show()
```

```python
labels = 'Generation X', 'Generation Z', 'Millenials'
sizes = [400639, 15906, 444261]
explode = [3, 0, 0]
plt.pie(sizes, explode=explode, labels=labels, 
autopct='%1.1f%%', shadow=False, radius= 100, startangle=-90, pctdistance=0.7)

plt.axis('equal')
plt.title('# of 5-24 Years Old Suicides of Each Generation ')
plt.show()
```
